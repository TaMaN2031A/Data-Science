```python
required_element = None

for i in soup.descendants:
    if i.getText().count("document.addEventListener(\"DOMContentLoaded\", function") != 0:
        required_element = i
```

```python
required_element
```

```python
type(required_element.getText())
```

```python
import re

match = re.search(r"var\s+viewModelData\s*=\s*(\[.*\]);", required_element.getText(), re.DOTALL)
json_str = match.group(1)
print(json_str[:500])
```

```python
import json

data = json.loads(json_str)
print(type(data))
print(data[0].keys())
```

```python
import pandas as pd
import re
from datetime import datetime

# دالة لتحويل التاريخ من الشكل /Date(1759438800000)/ إلى datetime
def parse_ms_date(s):
    if s is None:
        return None
    match = re.search(r'\d+', str(s))
    if not match:
        return None
    timestamp = int(match.group()) / 1000
    return datetime.utcfromtimestamp(timestamp)

# الكود الأساسي
rows = []

for day in data:
    day_date = parse_ms_date(day.get('Date'))
    matches = day.get('Matches', [])
    for match in matches:
        home = match.get('HomeTeamName')
        away = match.get('AwayTeamName')
        home_score = match.get('HomeScore')
        away_score = match.get('AwayScore')
        score = f"{home_score} - {away_score}" if home_score is not None and away_score is not None else "لم تُلعب بعد"

        rows.append({
            "اليوم": day_date.date() if day_date else None,
            "الفريق الأول": home,
            "الفريق الثاني": away,
            "أهداف الأول": int(home_score) if home_score is not None and away_score is not None else "لم تُلعب بعد",
            "أهداف الثاني": int(away_score) if home_score is not None and away_score is not None else "لم تُلعب بعد",
        })

df = pd.DataFrame(rows)
```
